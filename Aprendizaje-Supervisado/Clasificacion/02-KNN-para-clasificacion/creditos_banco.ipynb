```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import MinMaxScaler
from sklearn.neighbors import KNeighborsClassifier
```

```python
df = pd.read_csv('./creditos.csv')
df.head()
```

```python
buenos = df[df['cumplio']==1]
malos = df[df['cumplio']==0]
```

```python
plt.scatter(buenos['edad'], buenos['credito'],
            marker='*',s=150,color='skyblue',
            label='S pag (clase 1)')
plt.scatter(malos['edad'], malos['credito'],
            marker='*',s=150,color='red',
            label='No pag (clase 0)')
plt.xlabel('Monto del credito')
plt.ylabel('Edad')
plt.legend(bbox_to_anchor=(1,0.2));
```

```python
datos = df[['edad','credito']]
clase = df['cumplio']

escalador = MinMaxScaler()
datos = escalador.fit_transform(datos)
datos
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(datos,clase)
```

```python
edad = 40
monto = 350000

solicitante = escalador.transform([[edad,monto]])
print('clase ',knn.predict(solicitante))
print('probabilidades ',knn.predict_proba(solicitante))

plt.scatter(buenos['edad'], buenos['credito'],
            marker='*',s=150,color='skyblue',
            label='S pag (clase 1)')
plt.scatter(malos['edad'], malos['credito'],
            marker='*',s=150,color='red',
            label='No pag (clase 0)')
plt.scatter(edad, monto,
            marker='P',s=250,color='green',
            label='solicitante')
plt.xlabel('Monto del credito')
plt.ylabel('Edad')
plt.legend(bbox_to_anchor=(1,0.2));
```
