```python
import pandas as pd
```

```python
datos = pd.read_csv('ingreso.csv')
datos
```

```python
import matplotlib.pyplot as plt

plt.ylabel('Ingreso ($)')
plt.xlabel('Promedio de horas semanales trabajadas')
plt.scatter(datos['horas'],datos['ingreso'],color='pink');
```

### Creando el modelo de regresion

```python
from sklearn import linear_model

regresion = linear_model.LinearRegression()
horas = datos['horas'].values.reshape((-1,1))

modelo = regresion.fit(horas,datos['ingreso'])

print('Intercepcion (b)', modelo.intercept_)
print('Pendiente (m)', modelo.coef_)

entrada = [[39],[40],[43],[44]]

prediccion = modelo.predict(entrada)

plt.scatter(entrada, prediccion, color='red')
plt.plot(entrada,prediccion,color='black')

plt.ylabel('Ingreso ($)')
plt.xlabel('Promedio de horas semanales trabajadas')
plt.scatter(datos['horas'],datos['ingreso'],color='pink');
```
