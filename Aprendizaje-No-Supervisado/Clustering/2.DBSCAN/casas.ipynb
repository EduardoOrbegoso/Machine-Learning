```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
```

```python
df = np.loadtxt('./casas.csv',delimiter=',')
```

```python
clusters = DBSCAN(eps=2,min_samples=10).fit_predict(df)
clusters
```

```python
plt.figure(figsize=(7.5,7.5))
plt.scatter(df[:,0],df[:,1],c=clusters,s=100)
plt.xlabel('Antiguedad de la construccion en años')
plt.ylabel('Precio de las casas en pesos (1:100,000)')
plt.box(False);
```
